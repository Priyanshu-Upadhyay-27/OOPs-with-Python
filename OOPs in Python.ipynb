## Let's make a class of an ATM 
#### Where Data will be Balance and ATM pin 
#### Where Function will be withdrawl, deposit etc

**Constructor** is a special method, whose code execute automatically when the object of that class is created.

```python
class ATM:
    #Constructor
    def __init__(self):
        self.pin = ""  # All variables will be defined here
        self.balance = 0
        print(id(self))
        self.menu()   # Opened a menu when this object is created.
    
    def menu(self):
        user_input = input("""
                    Hello, how would you like to proceed?
                    1. Enter 1 to create pin
                    2. Enter 2 to deposit
                    3. Enter 3 to withdraw
                    4. Enter 4 to check balance
                    5. Enter 5 to exit
        """)

        if user_input == "1":
            self.create_pin()
        elif user_input == "2":
            self.deposit()
        elif user_input == "3":
            self.withdraw()
        elif user_input == "4":
            self.check_balance()
        else:
            print("Bye")
    
    def create_pin(self):
        self.pin = int(input("Enter you pin : "))
        print("Pin set successfully")
        
    def deposit(self):
        temp = int(input("Enter your pin : "))
        if temp == self.pin:
            amount = int(input("Enter the amount : "))
            self.balance = self.balance + amount
            print("Deposit Successfully")
        else:
            print("Invalid Pin")
    def withdraw(self):
        temp = int(input("Enter your pin : "))
        if temp == self.pin:
            amount = int(input("Enter the amount : "))
            if (self.balance - amount) >= 0:
                self.balance = self.balance - amount
                print("Debited Successfully")
            else:
                print("Insufficient Funds")
        else:
            print("Invalid Pin")
    def check_balance(self):
        temp = int(input("Enter your pin : "))
        if temp == self.pin:
            print(f"Your Savings Account have {self.balance} INR")
        else:
            print("Invalid Pin")
                
```

## Now, We will be creating our own datatype to handle fractions.

When user tries to print an instance of a class using print function, for eg. print(icici), then it tries to find the __str__ constructor and print the code inside it. If not present then just print the class information with memory location.

```python
class fraction:
    def __init__(self, n, d):
        self.num = n
        self.den = d
        
    def __str__(self):
        return f"{self.num}/{self.den}"
    
    def __add__(self, other): 
        # here two objects will be present, because 2 fractional number will be added. 
        # Now self contains the num and den for fraction 1 and other contains the num and den for fraction 2.
        # Here, 2 objects are used at once, therefore alone self is not sufficient. 
        temp_num = self.num * other.den + other.num * self.den
        temp_den = self.den * other.den 
        return f"{temp_num}/{temp_den}"
    def __sub__(self, other):
        temp_num = self.num * other.den - other.num * self.den
        temp_den = self.den * other.den 
        return f"{temp_num}/{temp_den}"
    def __mul__(self, other):
        temp_num = self.num * other.num
        temp_den = self.den * other.den
        return f"{temp_num}/{temp_den}"
    def __truediv__(self,other):
        temp_num = self.num / other.den
        temp_den = self.den / other.num
        return f"{temp_num}/{temp_den}"
```

### Encapsulation

Variable defined in constructor "init" inside a class are known as instance variable, because there values are unique with respect to each instance.

In python, you can hide the data defined in the constructor and some methods too, according to the needs, because think of icici bank, if someone uses the object, he or she can access the variables defined in the constructor and it would be bad, if someone can access the pin.

So, to keep the variable private, java uses private keyword, but python uses two underscore before a variable, meaning to make the pin variable private, it can be writtten as __pin. This syntax commands python to convert variable name **__pin** into **_(class name)__pin**, which makes the variable private and user cannot access it. So other people doesn't know, which variables are used and if they somehow know, the variables and if someone tries to change the pin : "self.__pin = 2187", then a new variable with __pin is created with value 2187. But since constructor has variables __pin, python converts all the __pin into _(class name)__pin and this is the variable which is containing the original value and not the __pin
    

But in python, everhting is not completely hidden,because if someone knows the variable and the name of the class then sbi._(class name)__pin [this syntax is common everywhere] can be used by them to view or change the original value. 

After hiding the data, we can make two methods, getter and setter, with the help of which we can get the pin and we can set the pin. These methods can beused by others, which rises a question that, why to create new methods if we want others to modify or view the instance variables?

So, the answer is we can control the way in which changes can be made when done using methods, but when done using class.instance_variable, others can do whatever they want without any checks, which can crash the code later. 

This is what we call **Encapsulation** i.e. adding protective barrier for the data and some methods so that, others cannot make unpredicted changes to the data or methods, thus maintaining data integrity and managing unauthorized changes.

```python
class ATM:
    #Constructor
    def __init__(self):
        self.__pin = ""  # All variables will be defined here
        self.__balance = 0
        print(id(self))
        self.menu()   # Opened a menu when this object is created.

    def get_pin(self):
        return self.__pin
    def set_pin(self, new_pin):
        self.__pin = new_pin
        print("Pin changes successfully")
    
    def menu(self):
        user_input = input("""
                    Hello, how would you like to proceed?
                    1. Enter 1 to create pin
                    2. Enter 2 to deposit
                    3. Enter 3 to withdraw
                    4. Enter 4 to check balance
                    5. Enter 5 to exit
        """)

        if user_input == "1":
            self.create_pin()
        elif user_input == "2":
            self.deposit()
        elif user_input == "3":
            self.withdraw()
        elif user_input == "4":
            self.check_balance()
        else:
            print("Bye")
    
    def create_pin(self):
        self.__pin = int(input("Enter you pin : "))
        print("Pin set successfully")
        
    def deposit(self):
        temp = int(input("Enter your pin : "))
        if temp == self.__pin:
            amount = int(input("Enter the amount : "))
            self.__balance = self.__balance + amount
            print("Deposit Successfully")
        else:
            print("Invalid Pin")
    def withdraw(self):
        temp = int(input("Enter your pin : "))
        if temp == self.__pin:
            amount = int(input("Enter the amount : "))
            if (self.__balance - amount) >= 0:
                self.__balance = self.__balance - amount
                print("Debited Successfully")
            else:
                print("Insufficient Funds")
        else:
            print("Invalid Pin")
    def check_balance(self):
        temp = int(input("Enter your pin : "))
        if temp == self.__pin:
            print(f"Your Savings Account have {self.__balance} INR")
        else:
            print("Invalid Pin")
                
```

## Pass By Refrence 

```python
class Customer:
    def __init__(self, name):
        self.name = name

def greet(customer):
    print(f"ID inside function: {id(customer)}")
    customer.name = "Ankita"  # Permanently changing the attribute
    print(f"Name inside function: {customer.name}")

# Main Code
cust = Customer("Nitesh")
print(f"ID outside function: {id(cust)}")
print(f"Name before function: {cust.name}")

greet(cust) # Both 'cust' and 'customer' point to the same address

print(f"Name after function: {cust.name}") # The name remains 'Ankita' because the object is mutable, therefore Nitish change to Ankita.
```

```python
def change(L):
    print(f"ID inside: {id(L)}")
    L.append(5) # Changes the original list

L1 = [1, 2, 3, 4]
print(f"ID outside: {id(L1)}")
change(L1)
print(f"Original list after function: {L1}") # Output: [1, 2, 3, 4, 5]
```

```python
def change_tuple(T):
    print(f"ID inside before change: {id(T)}")
    T = T + (5, 6) # Creates a NEW tuple; doesn't edit the original
    print(f"ID inside after change: {id(T)}") # Address will be different
    return T

T1 = (1, 2, 3, 4)
print(f"ID outside: {id(T1)}")
new_tuple = change_tuple(T1)
print(f"Original tuple after function: {T1}") # Output: (1, 2, 3, 4)
print(f"Memory Address of original tuple after function: {id(T1)}")
print(f"New tuple after function: {new_tuple}") # Output: (1, 2, 3, 4, 5, 6)
print(f"Memory Address of new tuple after function: {id(new_tuple)}")
```

These are the type of variable, which we can use commonly amoung different objects but of same class.

```python
class ATM:
    # 1. Static/Class variable: Defined inside the class but outside methods
    # Do not use ATM. prefix here during declaration
    __counter = 1 

    # Constructor
    def __init__(self):
        # 2. Instance variables: Defined using self
        # Two underscores make them private (Encapsulation)
        self.__pin = ""  
        self.__balance = 0
        
        # 3. Assigning the Serial Number
        # We access the class variable using the Class Name
        self.__sno = ATM.__counter
        ATM.__counter = ATM.__counter + 1

        print(f"Object created at memory address: {id(self)}")
        
        # 4. Open the menu ONLY after all data is initialized
        self.menu()

    # Static Methods: Deal with class-level data, no 'self' required
    @staticmethod
    def get_counter(): 
        return ATM.__counter

    @staticmethod
    def set_counter(new_val):
        if type(new_val) == int:
            ATM.__counter = new_val
        else:
            print("Not Allowed")
    
    # Instance Methods: Require 'self' to access object data
    def menu(self):
        user_input = input("""
                    Hello, how would you like to proceed?
                    1. Enter 1 to create pin
                    2. Enter 2 to deposit
                    3. Enter 3 to withdraw
                    4. Enter 4 to check balance
                    5. Enter 5 to exit
        """)

        if user_input == "1":
            self.create_pin()
        elif user_input == "2":
            self.deposit()
        elif user_input == "3":
            self.withdraw()
        elif user_input == "4":
            self.check_balance()
        else:
            print("Bye")
            return # Exit the method

    def create_pin(self):
        # Convert input to int to match the comparison in other methods
        self.__pin = int(input("Enter your pin: "))
        print("Pin set successfully")
        self.menu() # Go back to menu
        
    def deposit(self):
        temp = int(input("Enter your pin: "))
        if temp == self.__pin:
            amount = int(input("Enter the amount: "))
            self.__balance = self.__balance + amount
            print("Deposit Successfully")
        else:
            print("Invalid Pin")
        self.menu()
        
    def withdraw(self):
        temp = int(input("Enter your pin: "))
        if temp == self.__pin:
            amount = int(input("Enter the amount: "))
            if (self.__balance - amount) >= 0:
                self.__balance = self.__balance - amount
                print("Debited Successfully")
            else:
                print("Insufficient Funds")
        else:
            print("Invalid Pin")
        self.menu()

    def check_balance(self):
        temp = int(input("Enter your pin: "))
        if temp == self.__pin:
            print(f"Your Savings Account has {self.__balance} INR")
        else:
            print("Invalid Pin")
        self.menu()
```

### Aggregation

It is "HAS - A" relationship with different classes. Meaning, one class has a relationship with other i.e. one class has used another class as it's attribute. For Eg., Customer is the class which uses another complex class called address, which is unlike inheritance(smartphone is a product). ## Inheritance

**Inheritance** is the way through which, the child class inherits the properties(vairables, methods, constructor) of parent class. For eg., a course providing website(like udemy, coursera), has student and teacher which are required to login each time they visit the site. Since the login manner for both of them is same(using mail id and password), therefore, we can create a user class class which has the login and registration method and child classes i.e teacher and student can inherit these things from parent class.

```python
class user:
    def login(self):
        print("Login")
    def register(self):
        print("Register")

class student(user):  # this syntax ensures, user is the parent class and the child class student, inherits the login and register.
    def enroll(self):
        print("Enroll")
    def review(self):
        print("Review")
```

```python
class Parent:
    
    def __init__(self,num):
        self.__num=num
        
    def get_num(self):
        return self.__num
        
class Child(Parent):
    
    def show(self):
        print("This is in child class")
        
son=Child(100)
print(son.get_num())
son.show()
```

```python
class Parent:
    def __init__(self, num):
        self.__num = num

    def get_num(self):
        return self.__num

class Child(Parent):
    def __init__(self, val, num):
        self.__val = val

    def get_val(self):
        return self.__val

son = Child(100, 10)
print("Parent Num:", son.get_num()) # since parent constructor was never called and therefore, num variable was never defined
```

**Learning:** if a Child class defines its own init constructor, it effectively overrides the constructor of the Parent class.

### Super Keyword ###

This keyword helps in avoiding method overriding, and helps in calling the method from the parent class and not the method from child class. This super keyword cannot be used outside class. It can only be used to access parent's method and parent's constructor. Not even attributes can be accessed using super keyword. After the method from parent class gets executed completely, the control goes to the next line of the super keyword.

```python
class Phone:
    def __init__(self, price, brand, camera):
        print("Inside phone constructor")
        self.price = price
        self.brand = brand
        self.camera = camera
    def buy(self):
        print("Control comes here in the parent class")
        print("Buying a phone")
        
class SmartPhone(Phone):

    def buy(self):
        print("Buying a smartphone")
        print("Control goes to parent same name object")
        super().buy()
        print("Does it come back to here again?")
        

s = SmartPhone(20000, "Samsung", 10)
s.buy()
```

```python
class Phone:
    def __init__(self, price, brand, camera):
        print("Then, Inside phone constructor")
        self.price = price
        self.brand = brand
        self.camera = camera

class SmartPhone(Phone):
    def __init__(self, price, brand, camera, os, ram):
        # Calling parent constructor to handle shared attributes
        print("first you comes here")
        super().__init__(price, brand, camera)
        self.os = os
        self.ram = ram
        print("Then, Inside smartphone constructor")

s = SmartPhone(20000, "Samsung", 12, "Android", 2)
print(s.os)
print(s.brand)
```

```python
class Parent:
    def __init__(self, num):
        self.__num = num

    def get_num(self):
        return self.__num

class Child(Parent):
    def __init__(self, val, num):
        super().__init__(num)
        self.__val = val

    def get_val(self):
        return self.__val

son = Child(100, 200)
print("Child Val:", son.get_val())
print("Parent Num:", son.get_num()) 
# since parent constructor was called using super, therefore it has the value of nums
# due to inheritance, get_num() method can be called using the object of the child class.
```

### Types Of Inheritance

#### Single-level Inheritance

```python
class Phone:
    def __init__(self, price, brand, camera):
        print("Inside phone constructor")
        self.price = price
        self.brand = brand
        self.camera = camera

    def buy(self):
        print("Buying a phone")

class SmartPhone(Phone): # Inherits everything from Phone
    pass

s = SmartPhone(20000, "Apple", 13)
s.buy() # Output: Buying a phone
```

#### Multi-level Inheritance

```python
class Product:
    def review(self):
        print("Product review")

class Phone(Product): # Child of Product
    def buy(self):
        print("Phone buy")

class SmartPhone(Phone): # Child of Phone, Grandchild of Product
    pass

s = SmartPhone()
s.buy()    # Output: Phone buy
s.review() # Output: Product review
```

#### Hierarchial Inheritance

```python
class Phone:
    def __init__(self, price, brand, camera):
        self.price = price
        self.brand = brand
        self.camera = camera

    def buy(self):
        print("Buying a phone")

class SmartPhone(Phone): # First child
    pass

class FeaturePhone(Phone): # Second child
    pass

s = SmartPhone(10000, "Apple", 13)
f = FeaturePhone(5000, "Nokia", "2MP")
s.buy() # Both have access to the same parent method
f.buy()
```

#### Multiple Inheritance

```python
class Phone:
    def __init__(self, price, brand, camera):
        print("Inside phone constructor")
        self.price = price
        self.brand = brand
        self.camera = camera

    def buy(self):
        print("Phone buy")

class Product:
    def review(self):
        print("Customer review")

class SmartPhone(Phone, Product): # Inheriting from both, but first Phone class will be parsed for constructor or methods
    pass

s = SmartPhone(20000, "Apple", 12)
s.buy()    # Output: Phone buy
s.review() # Output: Customer review
```

**MRO(Method Resolution Order):** It occurs in multiple inheritance, which includes multiple parents to single child. If same methods exist in parent class, and they gets inherited to the child, so the method of first inherited parent class will be executed. |

For the constructor, the python first finds it in the child class, if not, then find it in the first parent class, if it finds then it executes the constructor first and stops the further search for the constructor.

### Method Overloading

Method Overloading is the ability to define a method in a way that it can be called with a different number or type of arguments.
It is not technically possible in python to perform method overloading,If you define a method twice, the second definition simply overwrites the first one. But it can be done in java. So we, make some modifications in the code to perform it other way.

```python
class Geometry:
    # We set 'b' to None by default
    def area(self, a, b=None):
        if b is not None:
            # If 'b' exists, it's a rectangle
            print("Area of Rectangle:", a * b)
        else:
            # If 'b' is None, it's a circle
            print("Area of Circle:", 3.14 * (a ** 2))

# Usage
obj = Geometry()
obj.area(4)       # Acts as a circle method
obj.area(4, 5)    # Acts as a rectangle method
```

```python
class Calculator:
    def add(self, *args):
        sum_val = 0
        for i in args:
            sum_val += i
        print("Total Sum:", sum_val)

# Usage
calc = Calculator()
calc.add(10, 20)           # Adding 2 numbers
calc.add(10, 20, 30, 40)   # Adding 4 numbers
```
